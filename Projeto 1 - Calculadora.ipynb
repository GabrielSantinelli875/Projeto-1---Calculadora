```python
print("calculadora do Gabriel")

while True: # looping de operações
    num1 = int(input("Digite o primeiro número: ")) # interação do usuário
    num2 = int(input("Digite o segundo número: ")) # interação do usuário

    print("\n Escolha uma operação:")
    print("+")
    print("-")
    print("*")
    print("/")

    operação_matemática = input("digite a operação desejada:") # variavel
if operação_matemática == "+": # condicionantes
    resultado = num1 + num2
    print("soma", resultado)
elif operação_matemática == "-":
    resultado = num1 - num2
    print("subtração", resultado)
elif operação_matemática == "*":
    resultado = num1 * num2
    print("multiplicação", resultado)
elif operação_matemática == "/":
    if num2 != 0:
      resultado = num1 / num2
      print("divisão", resultado)
    else:
      print("erro: nãao há divisão por zero")
else:
    print("opção invalida")
```
